### 3. Low Level Data Understanding

#### a. count of unique values in location column

```python
data['location'].value_counts()
```

#### b. Which continent has maximum frequency using values counts. 

```python
#m = max(data['continent'].value_counts())
count = data['continent'].value_counts().idxmax()
print('Continent which has maximun frequency is %s\n'%count)
```

#### c. Find maximun & mean value in 'total_cases'

```python
m = data['total_cases'].max()
print("Maximum value in 'total_cases' is %s"%m)
avg = data['total_cases'].mean()
print("Mean value in 'total_cases' is %s"%avg)
```

#### d. Find 25%, 50% & 75% quartile value in 'total_deaths'

```python
print("25%% quartile value in 'total_deaths' is %s"%data['total_deaths'].describe(include='all').loc['25%'])
print("50%% quartile value in 'total_deaths' is %s"%data['total_deaths'].describe(include='all').loc['50%'])
print("75%% quartile value in 'total_deaths' is %s"%data['total_deaths'].describe(include='all').loc['75%'])
```

#### e. Find which continent has maximum 'human_development_index'

```python
data.groupby(["continent"]).agg({"human_development_index":"max"})
```

#### f. Find which continent has minimum 'gdp_per_capita' 

```python
data.groupby(["continent"]).agg({"gdp_per_capita":"min"})
```
